```python
import pandas as pd
from ucimlrepo import fetch_ucirepo

mushroom = fetch_ucirepo(id=73)
mushroom_df = pd.DataFrame(data=mushroom.data.features)
mushroom_df.head()
```

```python
#AUGMENTACE DAT
mapping_cap_shape = {'b': 'bell', 'c': 'conical', 'x': 'convex', 'f': 'flat', 'k': 'knobbed', 's': 'sunken'}
mapping_cap_surface = {'f': 'fibrous', 'g': 'grooves', 'y': 'scaly', 's': 'smooth'}
mapping_cap_color = {'n': 'brown', 'b': 'buff', 'c': 'cinnamon', 'g': 'gray', 'r': 'green', 'p': 'pink', 'u': 'purple', 'e': 'red', 'w': 'white', 'y': 'yellow'}
mapping_bruises = {'t': 'bruises', 'f': 'no bruises'}
mapping_odor = {'a': 'almond', 'l': 'anise', 'c': 'creosote', 'y': 'fishy', 'f': 'foul', 'm': 'musty', 'n': 'none', 'p': 'pungent', 's': 'spicy'}
mapping_gill_attachement = {'a': 'attached', 'd': 'descending', 'f': 'free', 'n': 'notched'}
mapping_gill_spacing = {'c': 'close', 'w': 'crowded', 'd': 'distant'}
mapping_gill_size = {'b': 'broad', 'n': 'narrow'}
mapping_gill_color = {'k': 'black', 'n': 'brownn', 'b': 'buff', 'h': 'chocolate', 'g': 'gray', 'r': 'green', 'o': 'orange', 'p': 'pink', 'u': 'purple', 'e': 'red', 'w': 'white', 'y': 'yellow'}
mapping_stalk_shape = {'e': 'enlarging', 't': 'tapering'}
mapping_stalk_root = {'b': 'bulbous', 'c': 'club', 'u': 'cup', 'e': 'equal', 'z': 'rhizomorphs', 'r': 'rooted', '?': 'missing'}
mapping_stalk_surface_above_ring = {'f': 'fibrous', 'y': 'scaly', 'k': 'silky', 's': 'smooth'}
mapping_stalk_surface_below_ring = {'f': 'fibrous', 'y': 'scaly', 'k': 'silky', 's': 'smooth'}
mapping_stalk_color_above_ring = {'n': 'brown', 'b': 'buff', 'c': 'cinnamon', 'g': 'gray', 'o': 'orange', 'p': 'pink', 'e': 'red', 'w': 'white', 'y': 'yellow'}
mapping_stalk_color_below_ring = {'n': 'brown', 'b': 'buff', 'c': 'cinnamon', 'g': 'gray', 'o': 'orange', 'p': 'pink', 'e': 'red', 'w': 'white', 'y': 'yellow'}
mapping_veil_type = {'p': 'partial', 'u': 'universal'}
mapping_veil_color = {'n': 'brown', 'o': 'orange', 'w': 'white', 'y': 'yellow'}
mapping_ring_number = {'n': 'none', 'o': 'one', 't': 'two'}
mapping_ring_type = {'c': 'cobwebby', 'e': 'evanescent', 'f': 'flaring', 'l': 'large', 'n': 'none', 'p': 'pendant', 's': 'sheathing', 'z': 'zone'}
mapping_spore_print_color = {'k': 'black', 'n': 'brown', 'b': 'buff', 'h': 'chocolate', 'r': 'green', 'o': 'orange', 'u': 'purple', 'w': 'white', 'y': 'yellow'}
mapping_population = {'a': 'abundant', 'c': 'clustered', 'n': 'numerous', 's': 'scattered', 'v': 'several', 'y': 'solitary'}
mapping_habitat = {'g': 'grasses', 'l': 'leaves', 'm': 'meadows', 'p': 'paths', 'u': 'urban', 'w': 'waste', 'd': 'woods'}


mushroom_df['cap-shape'] = mushroom_df['cap-shape'].replace(mapping_cap_shape)
mushroom_df['cap-surface'] = mushroom_df['cap-surface'].replace(mapping_cap_surface)
mushroom_df['cap-color'] = mushroom_df['cap-color'].replace(mapping_cap_color)
mushroom_df['bruises'] = mushroom_df['bruises'].replace(mapping_bruises)
mushroom_df['odor'] = mushroom_df['odor'].replace(mapping_odor)
mushroom_df['gill-attachment'] = mushroom_df['gill-attachment'].replace(mapping_gill_attachement)
mushroom_df['gill-spacing'] = mushroom_df['gill-spacing'].replace(mapping_gill_spacing)
mushroom_df['gill-size'] = mushroom_df['gill-size'].replace(mapping_gill_size)
mushroom_df['gill-color'] = mushroom_df['gill-color'].replace(mapping_gill_color)
mushroom_df['stalk-shape'] = mushroom_df['stalk-shape'].replace(mapping_stalk_shape)
mushroom_df['stalk-root'] = mushroom_df['stalk-root'].replace(mapping_stalk_root)
mushroom_df['stalk-surface-above-ring'] = mushroom_df['stalk-surface-above-ring'].replace(mapping_stalk_surface_above_ring)
mushroom_df['stalk-surface-below-ring'] = mushroom_df['stalk-surface-below-ring'].replace(mapping_stalk_surface_below_ring)
mushroom_df['stalk-color-above-ring'] = mushroom_df['stalk-color-above-ring'].replace(mapping_stalk_color_above_ring)
mushroom_df['stalk-color-below-ring'] = mushroom_df['stalk-color-below-ring'].replace(mapping_stalk_color_below_ring)
mushroom_df['veil-type'] = mushroom_df['veil-type'].replace(mapping_veil_type)
mushroom_df['veil-color'] = mushroom_df['veil-color'].replace(mapping_veil_color)
mushroom_df['ring-number'] = mushroom_df['ring-number'].replace(mapping_ring_number)
mushroom_df['ring-type'] = mushroom_df['ring-type'].replace(mapping_ring_type)
mushroom_df['spore-print-color'] = mushroom_df['spore-print-color'].replace(mapping_spore_print_color)
mushroom_df['population'] = mushroom_df['population'].replace(mapping_population)
mushroom_df['habitat'] = mushroom_df['habitat'].replace(mapping_habitat)

mushroom_df.fillna('missing', inplace=True)

mushroom_df.head()




```

```python
from gensim.models import Word2Vec
sentences = []
for i in range(len(mushroom_df)):
    sentence = []
    for word in mushroom_df.iloc[i]:
        sentence.append(word)
    sentences.append(sentence)

print(len(sentences))
```

```python
import gensim
from gensim.models import Word2Vec

model = Word2Vec(sentences, min_count=1, vector_size=2, window=5)
print(model)

print(model.wv['smooth'])
```

```python
#VIZUALIZCE SLOV
import matplotlib.pyplot as plt

for word in model.wv.key_to_index:
    if word != 'nan' and isinstance(word, str):
        word_vector = model.wv[word]
        plt.scatter(x=word_vector[0], y=word_vector[1])
        plt.annotate(word, (word_vector[0], word_vector[1]))

plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.title("Word Embeddings Visualization")
plt.show()
```

```python
#VIZUALIZACE VET
import numpy as np

def sentence_embedding(sentence):
    word_embeddings = []
    for word in sentence:
        word_embeddings.append(model.wv[word])
    return np.mean(word_embeddings, axis=0)

# for sentence in sentences:
#     print(sentence_embedding(sentence))


for sentence in sentences:
    plt.scatter(x=sentence_embedding(sentence)[0], y=sentence_embedding(sentence)[1])

plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.title("Sentence Embeddings Visualization")
plt.show()
```
